##6. Box Plot (Seaborn)
o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

o Highlight the median and any outliers.

```python
import seaborn as sns  #importing
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 75]
median = np.median(marks) #storing the median value which is used to indicate the line of the median
plt.axhline(y=median, color='r', linestyle='--') #displaying the median line on the boxplot
q1, q3 = np.percentile(marks, [25, 75]) #to calculate the percentage of array which are values between 25 and 75
iqr = q3 - q1 #calculating iqr by the q3 & q1
lower_bound = q1 - 1.5 * iqr #formula for lower bound
upper_bound = q3 + 1.5 * iqr ##formula for upper bound
outliers = [x for x in marks if x < lower_bound or x > upper_bound] #to know  the which values are the outliers for the plot
sns.boxplot(x=marks, showmeans= True) #using the seaborn forplotting
if outliers:
  sns.scatterplot(x=outliers, y=[0] * len(outliers), color='violet') #if it is outlier then the color of the outlier is red
plt.show()
```

##8. Comparison of Matplotlib and Seaborn
o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.

o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.bar(categories, values, color='skyblue')
plt.xlabel("Categories")  #labeling
plt.ylabel("Values")  #labeling
plt.title("Bar Chart (Matplotlib)") #title
plt.show()

# Seaborn bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=values, palette="viridis") # Using a color palette
plt.xlabel("Categories")  #labeling
plt.ylabel("Values")  #labeling
plt.title("Bar Chart (Seaborn)") #title
plt.show()
```
